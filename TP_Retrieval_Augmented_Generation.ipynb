Chargement des documents

```python
import fitz  # PyMuPDF
import os

# Chemin vers le dossier contenant les fichiers PDF
data_path = "/content/tp-rag-student-version/data/arxiv"

# Fonction pour lire tous les fichiers PDF d'un dossier
def load_pdfs(directory):
    documents = []
    for filename in os.listdir(directory):
        if filename.endswith(".pdf"):
            filepath = os.path.join(directory, filename)
            with fitz.open(filepath) as pdf:
                text = ""
                for page in pdf:
                    text += page.get_text()
                documents.append({"text": text, "source": filename})
    return documents

# Charger les documents
documents = load_pdfs(data_path)
print(f"{len(documents)} documents chargés.")
```

Pré-traitement des documents

```python
# Fonction pour diviser un texte en morceaux de taille fixe
def split_text(text, chunk_size=500, overlap=50):
    chunks = []
    for i in range(0, len(text), chunk_size - overlap):
        chunks.append(text[i:i + chunk_size])
    return chunks

# Diviser les documents en paragraphes
split_documents = []
for doc in documents:
    chunks = split_text(doc["text"])
    for chunk in chunks:
        split_documents.append({"text": chunk, "source": doc["source"]})

print(f"{len(split_documents)} paragraphes créés.")
```

Calcul des embeddings

```python
from langchain.embeddings import HuggingFaceEmbeddings

# Charger le modèle d'embeddings
embeddings = HuggingFaceEmbeddings(model_name="intfloat/multilingual-e5-base")

# Générer les embeddings
embedded_documents = [
    {"embedding": embeddings.embed_query(doc["text"]), "source": doc["source"], "text": doc["text"]}
    for doc in split_documents
]

print(f"Embeddings calculés pour {len(embedded_documents)} paragraphes.")
```

Indexation dans un vector store

```python
from langchain.schema import Document

# Transformer les dictionnaires en objets Document
documents_as_objects = [
    Document(page_content=doc["text"], metadata={"source": doc["source"]})
    for doc in split_documents
]

# Indexer les embeddings dans Chroma
vectorstore = Chroma.from_documents(documents_as_objects, embedding=embeddings)

print("Indexation terminée avec succès.")
```

Exercice 2



1.Fonction d'interrogation



```python
# Fonction pour interroger la base vectorielle
def query_vectorstore(query, top_k=3):
    """
    Interroge la base vectorielle et retourne les documents les plus pertinents.

    :param query: La requête utilisateur (chaîne de caractères).
    :param top_k: Le nombre de documents à retourner (par défaut 3).
    :return: Liste des documents pertinents et leurs scores.
    """
    # Effectuer une recherche dans le vector store
    results = vectorstore.similarity_search_with_score(query, k=top_k)

    # Afficher les résultats
    for i, (doc, score) in enumerate(results):
        print(f"Document {i + 1} (Score : {score}):\n{doc.page_content}\n---")

    return results
```

2. Tester la fonction avec une requête

```python
# Exemple de requête
query = "Surfaces intelligentes reconfigurables"

# Interroger la base vectorielle
results = query_vectorstore(query, top_k=3)
```

1. Définir un prompt template

```python
# Exemple de contexte et question
example_context = "Les surfaces intelligentes reconfigurables améliorent les communications."
example_question = "Qu'est-ce qu'une surface intelligente ?"

# Générer le prompt complet
formatted_prompt = prompt_template.format(context=example_context, question=example_question)

print("Prompt formaté :")
print(formatted_prompt)
```
